### **Text Feature Extraction (Scikit-Learn)**

**1. Introduction**

In many machine learning tasks, the raw data is a collection of text documents. Since machine learning algorithms generally require numerical input, we must convert these text documents into numerical vectors. This process is called Feature Extraction.

In order to address this, the most common ways to extract numerical features from text content, namely:

- **Tokenizing** strings and giving integer id for each possible token, for instance by using white-spaces and punctuation as token separators

- **Counting** the occurrences of tokens in each document
- **Normalizing** and weighting with diminishing importance tokens that occur in the majority of samples/documents


In this scheme, **features** and **samples** are defined as follows:
- Each individual token occurrence frequency (normalized or not) is treated as a feature
- The vector of all the token frequencies for a given document is considered a multivariate sample


**2. The Bag of Words Representation**

A **corpus of documents** can thus be represented by a matrix with one row per document and one column per token (e.g. word) occurring in the corpus
- **Vectorization** is the general process of turning a collection of text documents into numerical feature vectors
- This specific strategy is called  the **Bag of Words** or “**Bag of n-grams**” representation  

We represent each document by the occurrence of words within it, ignoring the relative order of words.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample corpus
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Learn the vocabulary and transform the documents into vectors
X = vectorizer.fit_transform(corpus)

# View the learned features (vocabulary)
print("Feature Names:", vectorizer.get_feature_names_out())

# View the resulting array (Bag of Words)
print("Sparse Matrix Array:\n", X.toarray())
```

**3. Tokenizing and Preprocessing**

The **CountVectorizer** handles tokenization (splitting strings into words) and preprocessing (converting to lowercase).

- The default configuration tokenizes the string by extracting words of at least 2 letters.
-Each term found by the analyzer during the fit is assigned a unique integer index corresponding to a column in the resulting matrix.


```python
# To see which index corresponds to which word
print("Vocabulary Index Map:", vectorizer.vocabulary_)

# Encoding a new, unseen document
new_doc = ['Something completely new.']
print("New Document Vector:", vectorizer.transform(new_doc).toarray())
```

**4. Using Stop Words**

Stop words are common words (like "and", "the", "is") that often provide little information for distinguishing documents.

```python
# Re-initialize with English stop words
vectorizer_stop = CountVectorizer(stop_words='english')
X_stop = vectorizer_stop.fit_transform(corpus)

print("Features without stop words:", vectorizer_stop.get_feature_names_out())
```

**5. TF-IDF Term Weighting**

While counts are useful, frequent words across all documents (like "document" in our corpus) might be less informative than rare words. TF-IDF (Term Frequency-Inverse Document Frequency) scales the counts.

- TF = number of times a term **t** appears on document **d** / total number of term **t**  in document **d**
- IDF = total number of documents in corpus **D** / number of documents containing term **t**
- TF-IDF(t,d,D) = TF(t,d) x IDF(t, D)

```python
from sklearn.feature_extraction.text import TfidfTransformer

# Use the counts from our previous CountVectorizer
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)

print("TF-IDF Weighted Matrix:\n", tfidf.toarray())
```

**Shortcut: TfidfVectorizer**

You can perform both steps (counting and TF-IDF) in one go:

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(corpus)

print("TF-IDF Feature Names:", tfidf_vectorizer.get_feature_names_out())
```

**6. N-grams**

Sometimes, the order of words matters slightly (e.g., "not good" vs "good"). We can capture phrases using n-grams.

```python
# Extracting both individual words and pairs of words (unigrams and bigrams)
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))
X_ngram = ngram_vectorizer.fit_transform(corpus)

print("N-gram Features:", ngram_vectorizer.get_feature_names_out())
```
